```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

## Player Count

* Display the total number of players


```python
purchase_data.head()
```

```python
#Total Player    #Series.nunique(dropna=True) Return number of unique elements in the object.

total_players = purchase_data["SN"].nunique()
tot_players_dict={"Total Players":[total_players]}
tot_plyr_df = pd.DataFrame.from_dict(tot_players_dict, orient='columns')
tot_plyr_df

```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#number of unique items
unique = purchase_data["Item ID"].value_counts().count()

#need to figure out how to output as $0.00
#average purchase amount
avg = purchase_data["Price"].mean()

#Total Number of Purchases
total_purch = purchase_data["Purchase ID"].value_counts().count()

#need to figure out how to output as $0.00
#Total Revenue
Total_Revenue = purchase_data["Price"].sum()

p_dict = {"Number of Unique Items":[unique],
        "Average Price":[avg],
          "Number of Purchases":[total_purch],
          "Total Revenue":[Total_Revenue]}
p_df = pd.DataFrame.from_dict(p_dict, orient='columns')
p_df["Average Price"]= p_df["Average Price"].map("${:,.2f}".format)
p_df["Total Revenue"]= p_df["Total Revenue"].map("${:,.2f}".format)
p_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
unique_users_df = purchase_data.drop_duplicates(["SN"])
gender_group_df = unique_users_df.groupby(['Gender'])
gender_dict = {"Percentage of Players": gender_group_df["Gender"].count()/total_players*100,
               "Total Count": gender_group_df["SN"].count()}

gender_df = pd.DataFrame.from_dict(gender_dict,orient='columns')
gender_df = gender_df.sort_values("Total Count", ascending=False)


gender_df ["Percentage of Players"]= gender_df["Percentage of Players"].map("{:.2f}%".format)
gender_df.head()
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
gender_df = purchase_data.groupby(['Gender'])
purchase_dict = {"Purchase Count": gender_df["Purchase ID"].count(),
                 "Average Purchase Price": gender_df["Price"].mean(),
                "Total Purchase Value": gender_df["Price"].sum(),
                "Average Total Purchase per Person": gender_df["Price"].sum()/ gender_df["SN"].nunique()
                }
                 
purchase_df = pd.DataFrame.from_dict(purchase_dict,orient='columns')    

purchase_df["Average Purchase Price"]= purchase_df["Average Purchase Price"].map("${:,.2f}".format)
purchase_df["Total Purchase Value"]= purchase_df["Total Purchase Value"].map("${:,.2f}".format)
purchase_df["Average Total Purchase per Person"]= purchase_df["Average Total Purchase per Person"].map("${:,.2f}".format)

purchase_df

```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#### Create the bins in which Data will be held
  
bins = [0, 9,14,19,24,29,34,39,100]

# Create the names for the four bins
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

```

```python
purchase_data["Age_G"] = pd.cut(purchase_data["Age"], bins, labels=group_names)
```

```python
df = purchase_data.groupby("Age_G")
age_dict = {"Total Count":df["SN"].nunique(),
            "Percentage of Players":df["SN"].nunique()/total_players*100
           }
age_df = pd.DataFrame.from_dict(age_dict,orient='columns')
age_df["Percentage of Players"] = age_df["Percentage of Players"].map("{:.2f}%".format)

age_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
user_df = purchase_data.groupby("SN")
user_dict = {"Purchase Count":user_df["Purchase ID"].count(),
             "Average Purchase Price": user_df["Price"].mean(),
             "Total Purchase Value":user_df["Price"].sum() }
user_df2 = pd.DataFrame.from_dict(user_dict,orient='columns')
user_df2 = user_df2.sort_values("Total Purchase Value", ascending=False)

user_df2["Average Purchase Price"]= user_df2["Average Purchase Price"].map("${:,.2f}".format)
user_df2["Total Purchase Value"]= user_df2["Total Purchase Value"].map("${:,.2f}".format)
user_df2.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
reduced_purchase_data_df = purchase_data.loc[:, ["Item ID", "Item Name", "Price"]]

grouped_purchase_data_df = reduced_purchase_data_df.groupby(["Item ID", "Item Name"])

purch_dict = {"Purchase Count":grouped_purchase_data_df["Item ID"].count(),
             "Item Price": grouped_purchase_data_df["Price"].mean(),
             "Total Purchase Value":grouped_purchase_data_df["Price"].sum() }
purch_df2 = pd.DataFrame.from_dict(purch_dict,orient='columns')
purch_df2 = purch_df2.sort_values("Purchase Count", ascending=False)

purch_df2["Item Price"]= purch_df2["Item Price"].map("${:,.2f}".format)
purch_df2["Total Purchase Value"]= purch_df2["Total Purchase Value"].map("${:,.2f}".format)

purch_df2.head()

```
